```python
#AND gate implementation
import numpy as np
import tensorflow as tf

#example input 4 samples 2 featured each
X = tf.constant([[0,0],[0,1],[1,0],[1,1]],dtype = tf.float32)
```

```python
#manually define weights and bias
W = tf.constant([[2],[2]],dtype = tf.float32) #[[W-F1],[W-F2]]
b = tf.constant([-3],dtype = tf.float32)
```

```python
#forward pass: z = XW +b
z = tf.matmul(X,W) + b
```

```python
#act (step func / sigmoid)
y_step = tf.cast(z >=0, tf.int32) # hard thresh(binary output)
y_sigmoid = tf.math.sigmoid(z) # soft prob
```

```python
print("i/p: \n",X.numpy())
print("linear comb: \n",z.numpy())
print("step o/p perceptron: \n", y_step.numpy())
print("sig o/p: \n",y_sigmoid.numpy())
```
