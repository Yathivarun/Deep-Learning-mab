```python
#create slp for classification without using pre-defined libraries
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/students/iris.csv") #file location
```

```python
df.head()
```

```python
X = df.iloc[:,0:2].values
Y = df.iloc[:,-1].values
```

```python
y_binary = np.where(Y == "setosa",1.0,0.0).astype(np.float32) #where = condition, T=1,F=0
```

```python
X = (X - X.mean(axis = 0) / X.std(axis=0))
```

```python
X_tf = tf.constant(X ,dtype = tf.float32)
```

```python
W = tf.constant([[1,0],
                [-1,0]], dtype = tf.float32)
b = tf.constant([0.2], dtype = tf.float32 ) 
```

```python
z = tf.matmul(X_tf, W) + b
```

```python
X_tf.shape
```

```python
z.shape
```

```python
y_pred_step = tf.cast(z >= 0, tf.int32) #z>=0 o/p->1 else 0
```

```python
y_pred_sigmoid = tf.math.sigmoid(z)
```

```python
print("firt 10 true labels(setosa = 1, others = 0)",y_binary[:10])
```

```python
print("first 10 step o/p:",y_pred_step.numpy().flatten()[:10])
```

```python
print("first 10 sigmoid prob:",y_pred_sigmoid.numpy().flatten()[:10])
```

```python
#this program demonstrates how a SLP computes o/p with chosen weights(no W training) and bias
#step func -> classification(hard decision)
#sigmoid func -> probability interpretation
```

```python
for j,k in zip(y_binary,y_pred_step):
    if j==k:
        print("correct")
```
