```python
import numpy as np
```

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
```

```python
print("Tensor A:\n",A)
print("Tensor A:\n",B)
```

```python
#basic operations
add_res = A + B
sub_res = A - B
mul_res = A * B
div_res = A / B
matmul_res = A @ B
print("\n Add: ",add_res)
print("\n sub: ",sub_res)
print("\n mul: ",mul_res)
print("\n div: ",div_res)
print("\n manual: ",matmul_res)
```

```python
import tensorflow as tf

A = tf.constant([[1,2],[3,4]],dtype = tf.float32)
B = tf.constant([[1,2],[3,4]],dtype = tf.float32)
```

```python
#baSIC OPERATIONS
```

```python
add_res = tf.add(A,B)
sub_res = tf.subtract(A,B)
mul_res = tf.multiply(A,B)
div_res = tf.divide(A,B)
matmul_res = tf.matmul(A,B)
```

```python
print("add_res",add_res.numpy())
print("sub_res",sub_res.numpy())
print("mul_res",mul_res.numpy())
print("div_res",div_res.numpy())
print("matmul_res",matmul_res.numpy())
```
